# <span style="color: red;"> 크기 단위  </span>
## **px**: 화소 1개(디스플레이 상의 점하나) ex) font-size: 14px;
## **%**: 디폴트 사이즈의 지정한 비율로 설정  ex) font-size: 120%; ->디폴트값*1.2
## **em**: 디포틀에 배수 단위로 설정  ex)font-size: 1.2em;  -> 디폴트값에 1.2배
####   &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;          이때 여러자식들도 같이 적용될경우 1->1.2->1.44->1.96으로 후손으로 갈수록 커질수 있음
## **rem**(root em): 그래서 이거 사용->루트인 html값이 절대적인 기준  ex)font-size: 1.2rem;
## **viewport**: 브라우전 화면상 영역에 대한 단위
#### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;vw(width), vh(heigth), vmin, vmax등

# <span style="color: blue;">2.표현방식 </span>
## **display**: block, inline, inline-block, none
  
- block: 항상 새로운 라인에서 시작 ex) div, h1~h6, p, ol, ul, li, hr, table, form
- inline: 문장 중간에 들어가기 가능 ex) span, a, strong, img, br, input, select, button, textarea
- inline-block: 두가지 특성 다가짐

## **visibility**: visible(기본값), hidden(공간은 남음), collapse(행, 열 안보이게), none(공간까지 사라짐)

## **opacity**: 0~1사이의 값, 0=투명, 1=불투명

# <span style="color: blue;"> 5.위치  </span>
## -**position**과 **top, bottom, left, right** 같이 사용해서 콘텐츠 위치 이동
## **position:** static(고정) or relative or absolute or fixed
## 이후 **top:** 50px; **left:** 50px;등처럼 기존 위치에서 수치만큼 이동

## **z-index**: 지정한 숫자값이 클수록 앞에 위치
## **overflow:** visible or hidden or scroll or auto  -범위를 벗어났을때 어떻게 표시될지
## **float:** left(왼쪽정렬) or right(오른쪽정렬)
